```python
from bs4 import BeautifulSoup as bs
```

```python
from urllib.request import urlopen
```

```python
url='https://wuzzuf.net/search/jobs/?q=illustrator&a=navbl'
```

```python
client=urlopen(url)
```

```python
html=client.read()
```

```python
client.close()
```

```python
soup=bs(html,"html.parser")
```

```python
soup
```

```python
containers=soup.find_all('div',{'class':'css-1gatmva e1v1l3u10'})
```

```python
len(containers)
```

```python
bs.prettify(containers[0])
```

```python
containers[0].div.h2.text
```

```python
job_title=containers[0].findAll("h2",{'class':'css-m604qf'})
job_title[0].text
```

```python
company_name=containers[0].findAll("a",{'class':'css-17s97q8'})
company_name[0].text
```

```python
job_type=containers[0].findAll("span",{'class':'css-1ve4b75 eoyjyou0'})
job_type[0].text
```

```python
f=open('data/wuzzuf-illustrator.csv','w')
```

```python
header="job_title,company_name,job_type\n"
f.write(header)
```

```python
for container in containers:
    jtitle=container.findAll("h2",{'class':'css-m604qf'})
    job_title=jtitle[0].text.strip()
    cname=container.findAll("a",{'class':'css-17s97q8'})
    company_name=cname[0].text.strip()
    jtype=container.findAll("span",{'class':'css-1ve4b75 eoyjyou0'})
    job_type=jtype[0].text
    #print(job_title+','+company_name+','+job_type)
    f.write(job_title+','+company_name+','+job_type+'\n')
f.close()
```
